```python
data_cleaned=data.copy()
```

```python
#Data cleaning and pre-processing steps
columns_to_drop = ['img', 'release_date', 'last_update']
data_cleaned = data_cleaned.drop(columns=columns_to_drop)
```

```python
text_columns = ['title', 'console', 'genre', 'publisher', 'developer']
for col in text_columns:
    data_cleaned[col] = data_cleaned[col].fillna('Unknown')
```

```python
sales_columns = ['total_sales', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales']
for col in sales_columns:
    data_cleaned[col] = pd.to_numeric(data_cleaned[col], errors='coerce')
    median = data_cleaned[col].median()
    data_cleaned[col] = data_cleaned[col].fillna(median)
```

```python
data_cleaned['critic_score'] = pd.to_numeric(data_cleaned['critic_score'], errors='coerce')
median_score = data_cleaned['critic_score'].median()
data_cleaned['critic_score'] = data_cleaned['critic_score'].fillna(median_score)
data_cleaned['critic_score'] = data_cleaned['critic_score'].clip(0, 100)
```

```python
data_cleaned = data_cleaned.drop_duplicates()
```

```python
#Print the null values from all the columns of the cleaned data
print(data_cleaned.isnull().sum())
```

```python
#Encoding the Object datatype columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
columns_to_encode = ['title', 'console', 'genre', 'publisher', 'developer']
for col in columns_to_encode:
    data_cleaned[col] = le.fit_transform(data_cleaned[col])
```

```python
print(data_cleaned.info())
```

```python
#Outlier detection and removal using the Z-Score normalization method
import numpy as np
import pandas as pd
from scipy.stats import zscore
z_scores = np.abs(zscore(data_cleaned.select_dtypes(include=[np.number])))  # Only numeric columns
threshold = 3
outliers = (z_scores > threshold).any(axis=1)
df_cleaned = data_cleaned[~outliers]
print(df_cleaned)
```

```python
#Sequential data modelling for time series forecasting
def create_supervised_data(df_cleaned, columns, time_steps):
    X, y = [], []
    for i in range(time_steps, len(df_cleaned)):
        X.append(df_cleaned[columns].iloc[i-time_steps:i].values)
        y.append(df_cleaned['total_sales'].iloc[i])
    return np.array(X), np.array(y)

time_steps = 60
columns = ['genre', 'critic_score', 'na_sales', 'jp_sales', 'pal_sales', 'other_sales']
X, y = create_supervised_data(df_cleaned, columns, time_steps)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("\nThe dimension of the training dataset are:")
print(X_train.shape)

print("\nThe dimensions of the test dataset are:")
print(X_test.shape)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score

# Create and train the SVR model
svr_model = SVR(kernel='rbf')
svr_model.fit(X_train.reshape(X_train.shape[0], -1), y_train)

# Predict on the test data
y_pred = svr_model.predict(X_test.reshape(X_test.shape[0], -1))

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error (SVR): {mae}')

mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (SVR): {mse}')

r2 = r2_score(y_test, y_pred)
print(f'R^2 Score (SVR): {r2}')

evs = explained_variance_score(y_test, y_pred)
print(f'Explained Variance Score (SVR): {evs}')
```

```python
import matplotlib.pyplot as plt

# Plot Actual vs Predicted
plt.figure(figsize=(20, 10))
plt.plot(y_test, color='blue', label='Actual Values')
plt.plot(y_pred, color='red', linestyle='--', label='Predicted Values')
plt.title('Support Vector Machine (SVM) Forecasting', fontsize=16)
plt.xlabel('Time Index', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score

# Create and train the RandomForest model
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train.reshape(X_train.shape[0], -1), y_train)

# Predict on the test data
y_pred_rf = rf_model.predict(X_test.reshape(X_test.shape[0], -1))

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred_rf)
print(f'Mean Absolute Error (Random Forest): {mae}')

mse = mean_squared_error(y_test, y_pred_rf)
print(f'Mean Squared Error (Random Forest): {mse}')

r2 = r2_score(y_test, y_pred_rf)
print(f'R^2 Score (Random Forest): {r2}')

evs = explained_variance_score(y_test, y_pred_rf)
print(f'Explained Variance Score (Random Forest): {evs}')
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Prepare the data
y = data_cleaned['total_sales']

# Train-test split
train_size = int(len(y) * 0.75)
train, test = y[:train_size], y[train_size:]

# Create and fit the ARIMA model
model = ARIMA(train, order=(3, 0, 2))
model_fit = model.fit()

# Forecasting
y_pred_arima = model_fit.forecast(steps=len(test))

# Evaluate the model
mae = mean_absolute_error(test, y_pred_arima)
print(f'Mean Absolute Error (ARIMA): {mae}')

mse = mean_squared_error(test, y_pred_arima)
print(f'Mean Squared Error (ARIMA): {mse}')

r2 = r2_score(test, y_pred_arima)
print(f'R^2 Score (ARIMA): {r2}')

evs = explained_variance_score(test, y_pred_arima)
print(f'Explained Variance Score (ARIMA): {evs}')
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Prepare the data
y = data_cleaned['total_sales']

# Train-test split
train_size = int(len(y) * 0.75)
train, test = y[:train_size], y[train_size:]

# Create and fit the SARIMAX model
model = SARIMAX(train, order=(3, 0, 2), seasonal_order=(3, 0, 2, 12))
model_fit = model.fit()

# Make predictions
y_pred_sarimax = model_fit.forecast(steps=len(test))

# Evaluate the model
mae = mean_absolute_error(test, y_pred_sarimax)
print(f'Mean Absolute Error (SARIMA): {mae}')

mse = mean_squared_error(test, y_pred_sarimax)
print(f'Mean Squared Error (SARIMA): {mse}')

r2 = r2_score(test, y_pred_sarimax)
print(f'R^2 Score (SARIMA): {r2}')

evs = explained_variance_score(test, y_pred_sarimax)
print(f'Explained Variance Score (SARIMA): {evs}')
```
